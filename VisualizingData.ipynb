```python
#calculate the amplitude
import numpy as np
def Ampli(signal , frame_size):
    ampli = []
    for i in range(0,len(signal),frame_size):
        current_frame_ampli = max(signal[i:i+frame_size])
        ampli.append(current_frame_ampli)
    return np.array(ampli)
#here frame_size is the second variable we are geting after loading audio files
#Basically it is sampling rate 
#we are taking max value in each sample rate to get amplitude 
```

```python
Ampli_pos = Ampli(x_pos,Sr_pos)
```

```python
len(Ampli_pos)
```

```python
Ampli_neg = Ampli(x,Sr)
```
